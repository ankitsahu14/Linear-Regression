```python
def vif_value_calc(input_data_passed , dependent_column):
    vif_dataframe = pd.DataFrame( columns = ['Var', 'Vif'])
    x_variables=input_data_passed.drop([dependent_column], axis=1)
    xvariable_names=x_variables.columns
    for i in range(0,xvariable_names.shape[0]):
        y=x_variables[xvariable_names[i]] 
        x=x_variables[xvariable_names.drop(xvariable_names[i])]
        r_squared_value=sm.OLS(y,x).fit().rsquared  
        vif_value=round(1/(1-r_squared_value),2)
        vif_dataframe.loc[i] = [xvariable_names[i], vif_value]
    return vif_dataframe.sort_values(by = 'Vif', axis=0, ascending=False, inplace=False)
```

## Assesing the model

Let us plot the error items and see if it is just random noise or it has some definite pattern. 

```python
#Plotting the error items

figr = plt.figure()
c1 = [i for i in range(1,63,1)]
plt.scatter(c1, y_test-y_prediction_model9)
figr.suptitle('Error Items' , fontsize= 25)
plt.xlabel('Index', fontsize= 15)
plt.ylabel('y_test-y_pred', fontsize= 15)
```

```python
#To understand the distribution of the error items we need to use distplot

figr= plt.figure()
sns.distplot((y_test-y_prediction_model9), bins=20)
figr.suptitle('Error Items' , fontsize= 25)
plt.xlabel('Index', fontsize= 15)
plt.ylabel('y_test-y_pred', fontsize= 15)
```

We can see in this plot that the error items are randomly distributed and do not follow any pattern. We should therefore be good to go ahead with this model. The values of R Squared and Adjusted R Square is both close to 0.89
